```python
import pandas as pd
import json
```

```python
basic_stats = None
with open('./tree_stats_results.json') as file:
    basic_stats = json.loads(file.read().rstrip())
```

```python
basic_stats.keys()
```

```python
for stat in basic_stats.keys():
    stat_df = pd.DataFrame.from_dict(basic_stats[stat], orient='index', columns=['count'])
    # cast and sort
    stat_df.index = stat_df.index.astype('int64')
    stat_df.sort_index(inplace=True)

    stat_df.plot(y='count', logy=True, title=stat, xlabel='stat_value', ylabel='count')
    
```

```python
chain_stats = None
with open('./chain_results.json') as file:
    chain_stats = json.loads(file.read().rstrip())
    
```

```python
for leaning in chain_stats.keys():
    stat_df = pd.DataFrame.from_dict(chain_stats[leaning], orient='index', columns=['visitor_percent'])
    # cast and sort
    stat_df.index = stat_df.index.astype('int64')
    stat_df.sort_index(inplace=True)

    stat_df.plot(y='visitor_percent', title=leaning + ' video', xlabel='chain_length', ylabel='% comments from visitors')
    
```

```python
# overlay the two graphs with two colors
# theoretical lower bound
```

```python
lower_bound_df = pd.DataFrame({'chain_length': [i for i in range(1,285)],
                              'visitor_percent': [1.0 / i for i in range(1,285)],
                              'leaning': ['lower bound' for i in range(1,285)]})
```

```python
chain_stats_list = []
chain_stats_list.append(lower_bound_df)
```

```python
for leaning in chain_stats.keys():
    stat_df = pd.DataFrame.from_dict(chain_stats[leaning], orient='index', columns=['visitor_percent'])
    # cast and sort
    stat_df.index = stat_df.index.astype('int64')
    stat_df.sort_index(inplace=True)
    stat_df.reset_index(inplace=True)
    stat_df.rename(columns={'index': 'chain_length'}, inplace=True)
    stat_df['leaning'] = [leaning for i in range(len(stat_df))]
    chain_stats_list.append(stat_df)

df = pd.concat(chain_stats_list)
```

```python
df.head()
```

```python
df.groupby('leaning')['visitor_percent'].plot(legend=True, xlabel='chain_length', ylabel='% comments from visitors')
```
